```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np
import math
import sys

# time change (discrete time)
T = 365.0 #days
dt = 0.04
#time array
t = np.linspace(0, T, int(T/dt))



# array  to store  the  solution if you were using Euler's method
N = np.zeros((len(t)))
P1 = np.zeros((len(t)))
P2 = np.zeros((len(t)))
C = np.zeros((len(t)))
N[0] = 1
P1[0] = 0.5
P2[0] = 0.5
C[0] = 0.5

mu1 = np.empty((len(t)))
mu2 = np.empty((len(t)))
phi1 = np.empty((len(t)))
phi2 = np.empty((len(t)))

#initial light irradiance (constant)
I_naughts = np.empty(len(t))

#inital light irradiance (seasonal)
I_max = 24
I_min = 4

i = 0
for a in t:
    if a>1095.0:
        I_naughts[i] = I_min
    else:
        I_naughts[i] = I_min + (I_max-I_min)/2.*(np.sin(((a)/365*2*np.pi-np.pi/2))+1)
    i = i + 1
    
i = 0   
for a in t:
    #if a>1095.0:
    mu2[i] = (I_max/12)
    #else:
        #mu2[i] = np.absolute(I_naughts[i])/16.*(np.sin(((a)/365*2*np.pi-np.pi/2))+1)
    i = i + 1

i = 0    
for a in t:
    #if a>1095.0:
    phi2[i] = (I_max/240)
    #else:
       # phi2[i] = np.absolute(I_naughts[i])/70.*(np.sin(((a)/365*2*np.pi-np.pi/2))+1)
    i = i + 1

i = 0   
for a in t:
    #if a>1095.0:
    mu1[i] = (I_max/8)
    #else:
        #mu1[i] = np.absolute(I_naughts[i])/12.*(np.sin(((a)/365*2*np.pi-np.pi/2))+1)
    i = i + 1

i = 0    
for a in t:
    #if a>1095.0:
    phi1[i] = (I_max/120)
    #else:
        #phi1[i] = np.absolute(I_naughts[i])/60.*(np.sin(((a)/365*2*np.pi-np.pi/2))+1)
    i = i + 1
```

```python
plt.plot(t, I_naughts,color = 'black')
#plt.plot(t, mu1, color = 'orange')
#plt.plot(t, mu2, color = 'blue')
#plt.plot(t, phi1,'--', color = 'orange')
#plt.plot(t, phi2,'--', color = 'blue')
plt.title('Temperature Fluctuation')
plt.xlabel('Days')
plt.ylabel('Temperature (Degrees Celcius)')

plt.savefig('temperature_fluctuation.png', bbox_inches='tight', dpi=300)
```

```python
# values from tables.docx
#phi1 = 0.2 #interaction strength for p1
#phi2 = 0.1 #interaction strength for p2
#mu1 = 3.0 #resource affinity parameter for p1
#mu2 = 2.0 #resource affinity parameter for p2
eps1 = 0.4 #transfer efficiency for p1
eps2 = 0.3 #transfer efficiency for p2
delta1 = 0.1 #producer mortality
delta2 = 0.1 #producer mortality
deltaC = 0.1 #consumer mortality
deltaN = 0.01 #nutrient degredation


S_R = 1 #resource supply

#mu & phi will be functions of temperature
#linear relationship with temperature




```

```python
for i in range(1, len(t)):
        #mu1 = I_naughts[i]/12
        #phi1 = I_naughts[i]/40
        #mu2 = I_naughts[i]/16
        #phi2 = I_naughts[i]/40
        dNdt = S_R - (mu1[i]*N[i-1]*P1[i-1]) - (mu2[i]*N[i-1]*P2[i-1]) - deltaN*N[i-1]
        dP1dt = (mu1[i]*N[i-1]*P1[i-1]) - (phi1[i]*P1[i-1]*C[i-1]) - (delta1*P1[i-1])
        dP2dt = (mu2[i]*N[i-1]*P2[i-1]) - (phi2[i]*P2[i-1]*C[i-1]) - (delta2*P2[i-1])
        dCdt = (eps1*phi1[i]*P1[i-1]*C[i-1]) + (eps2*phi2[i]*P2[i-1]*C[i-1]) - (deltaC*C[i-1])
        N[i] = N[i-1] + dNdt * dt
        P1[i] = P1[i-1] + dP1dt * dt
        P2[i] = P2[i-1] + dP2dt * dt
        C[i] = C[i-1] + dCdt * dt
```

```python
plt.plot(t,N, color ='red')
plt.plot(t,P1, '--', color = 'orange')
plt.plot(t,P2, 'blue')
plt.plot(t,C, linestyle = 'dotted', color='gray')
plt.legend(['Nitrogen','Producer 1','Producer 2', 'Consumer'], loc ='upper right')
plt.xlabel('Days')
plt.ylabel('Producer/Consumer Biomass \n& Nitrogen Concentration \n (mmol C/ m$^{3}$)')
plt.title('High Temperature with Trade-off')

plt.savefig('temp_biomass.png', bbox_inches='tight', dpi=300)

plt.show()
```
