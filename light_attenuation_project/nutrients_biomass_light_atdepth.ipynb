```python
from scipy.integrate import *
from pandas import *
from numpy import *
from pylab import *
from scipy import *
import math
```

```python
alpha_knot = 1
z = 0 #depth, m
k = 0.04 #coefficient, m^-1
alpha = alpha_knot * math.exp(-k*z)
print(alpha)
```

```python
# time  step
dt = 0.04

# time change (discrete time)
T = 205 / 24.0
t = np.linspace(0, T, 2000)
#t = np.linspace(0,T,int(T/dt))
#print(t)

# array  to store  the  solution
N = np.zeros(len(t))
P = np.zeros(len(t))
N[0] = 2
P[0] = 2
y0 = N[0], P[0]
print(y0)
```

```python
def nbl_model(y, t, alpha):
    N,P = y[0],y[1]
    dNdt = -alpha*N*P
    dPdt = alpha *N*P
    return dNdt, dPdt
```

```python
integ_sivz = odeint(nbl_model, y0, t, args=(alpha,))
print(integ_sivz)
plt.plot(t,integ_sivz[:,0],'r') #nutrients
plt.text(4,1.75,'depth = 0 meter')
```

```python
plt.plot(t,integ_sivz[:,1],'b') #biomass
plt.text(0,3.75,'depth = 0 meter')
```
