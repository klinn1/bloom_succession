```python
# time change (discrete time)
T = 360.0 
delt = 1.0/240.0
t = np.linspace(0, T, 50) #time array
print(t.shape)
```

```python
shape = 50
value = 30


P1 = np.empty(shape, dtype=np.int)
P1.fill(value)
```

```python
zetas = np.linspace(0,100,50)
k_w = 0.02
k_p = 0.001
print(len(zetas))
```

```python
fig, ax = plt.subplots()
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False

ax.plot(P1,zetas)
ax.invert_yaxis()
ax.set_ylabel('Depth (m)')
ax.set_title('Biomass')
plt.show()
```

```python
I_naughts = (np.sin(((t)/T*2*math.pi-math.pi/2))+1)/2
#alpha_knots = ones(int(T/delt))
plt.plot(t,I_naughts)
plt.show()
```

```python
print(len(I_naughts))
print(I_naughts)
```

```python
def irradiance_function(zetas, P1, i, j, dep_size):
    I = I_naughts[i] * np.exp(-(k_w*zetas[j]+np.sum(k_p*P1[:j]*dep_size)))
    return I
```

```python
j = 0
i = 25
dep_size = 2

light_list = []
for j in range(len(zetas)):
    loop = irradiance_function(zetas, P1, i, j, dep_size)
    light_list.append(loop)
    
light_array = np.array(light_list)
```

```python
fig, ax = plt.subplots()
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False

ax.plot(light_array,zetas)
ax.invert_yaxis()
ax.set_ylabel('Depth (m)')
ax.set_xlabel('Biomass')
plt.show()

#two plots
#control - no biomass
#plot with biomass
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,3))
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False

axs[0].plot(light_array,zetas, color = 'blue')
axs[0].invert_yaxis()
axs[0].set_ylabel('Depth (m)')
axs[0].set_xlabel('Irradiance')
axs[0].set_title('With Biomass')


P2 = np.zeros(50)
P2 = -np.sort(-P2)
print(len(P2))
print(P2)
i = 25
dep_size = 2

light_list1 = []
for j in range(len(zetas)):
    loop1 = irradiance_function(zetas, P2, i, j, dep_size)
    light_list1.append(loop1)
    
light_array1 = np.array(light_list1)

axs[1].plot(light_array1,zetas, color = 'red')
axs[1].invert_yaxis()
axs[1].set_ylabel('Depth (m)')
axs[1].set_xlabel('Irradiance')
axs[1].set_title('No Biomass')

plt.show()
```
