```python
from scipy.integrate import *
from pandas import *
from numpy import *
from pylab import *
from scipy import *
import math
from scipy.interpolate import InterpolatedUnivariateSpline
```

```python
# values from tables.docx
phi1 = 0.001 #interaction strength for p1
phi2 = 0.001 #interaction strength for p2
eps1 = 0.1 #transfer efficiency for p1
eps2 = 0.1 #transfer efficiency for p2
m_c = 0.1 #consumer mortality
```

```python
# time change (discrete time)
T = 360.0 
delt = 1.0/240.0
t = np.linspace(0, T, int(T/delt)) #time array
#t = np.linspace(0,T,int(T/dt))
#print(t)

# array  to store  the  solution
N = np.zeros(len(t))
P1 = np.zeros(len(t))
P2 = np.zeros(len(t))
C = np.zeros(len(t))
N[0] = 500
P1[0] = 100
P2[0] = 100
C[0] = 100
y0 = np.array([N[0], P1[0],P2[0],C[0]]) #create an array with your starting values for odeint
print(y0)
```

```python
zetas = np.linspace(0,100,50)
k_w = 0.01
k_p = 0.01
print(len(zetas))
```

```python
I_naughts = (np.sin(((t)/T*2*math.pi-math.pi/2))+1)/2
#alpha_knots = ones(int(T/delt))
plt.plot(t,I_naughts)
```

```python
Is = []
for I_naught in I_naughts:
    I = I_naught * np.exp((-k_w*zetas)-(k_p*(P1[0]+P2[0])))
    Is.append(I)
light_att = np.array(Is)
lights = light_att.T

print(zetas.shape)
print(lights.shape)
```

```python
print(lights.shape)
```

```python
plt.contourf(t,zetas,lights,cmap = 'plasma')
plt.gca().invert_yaxis()
plt.xlabel('Time (Days)', color = 'k')
plt.ylabel('Depth (Meters)', color = 'k')
plt.title('Light Intensity')
plt.colorbar()
```

```python
I_naughts_interp = InterpolatedUnivariateSpline(t, I_naughts)
```

```python
plt.plot(t, I_naughts)
plt.plot(t, I_naughts_interp(t), 'r')
plt.show()
```

```python
def npc_model(y, t, zeta, delta, m_c, phi1, phi2, eps1, eps2, mu1, mu2):
    N, P1, P2, C = y[0], y[1], y[2], y[3]
    I_naught = I_naughts_interp(t)
    if I_naught < 0:
        I_naught = 0.
    I = I_naught * np.exp((-k_w*zeta))
    mu1_growth = np.minimum((N/(N+(mu1/alpha_n))), (I/(I+(mu1/alpha_i))))
    mu2_growth = np.minimum((N/(N+(mu2/alpha_n))), (I/(I+(mu2/alpha_i))))
    dNdt = -mu1_growth*P1 - mu2_growth*P2 + delta*P1 + delta*P2
    dP1dt = mu1_growth*P1 - delta*P1 - phi1*P1*C
    dP2dt = mu2_growth*P2 - delta*P2 - phi2*P2*C
    dCdt = eps1*phi1*P1*C + eps2*phi2*P2*C - m_c*C
    return np.array([dNdt, dP1dt, dP2dt, dCdt])

#np.minimum((N/(N+(mu1/alpha_n))), (I/(I+(mu1/alpha_i))))
#np.minimum((N/(N+(mu2/alpha_n))), (I/(I+(mu2/alpha_i))))
#-k_p*(P1+P2)
```

```python
delta = 0.12
mu1 = 1
mu2 = 1
alpha_n = 1
alpha_i = 1
tim = []
for zeta in zetas:
    integ_sivz = odeint(npc_model, y0, t, args=(zeta, delta, m_c, phi1, phi2, eps1, eps2, mu1, mu2))
    tim.append(integ_sivz)
sols = array(tim)
print(sols)
```

```python
plt.contourf(t,zetas,sols[:,:,0],cmap = 'plasma')
plt.gca().invert_yaxis()
plt.xlabel('Time (Days)', color = 'k')
plt.ylabel('Depth (Meters)', color = 'k')
plt.title('Nutrients')
plt.colorbar()
```
