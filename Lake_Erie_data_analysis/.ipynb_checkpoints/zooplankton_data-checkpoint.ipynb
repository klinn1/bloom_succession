```python
year = lake_erie_zooplankton['YEAR']
month = lake_erie_zooplankton['Month']
calanoid_adults = lake_erie_zooplankton['CAL_ugDW/m3']
calanoid_babies = lake_erie_zooplankton['CALIM_ugDW/m3']
cyclopoid_adults = lake_erie_zooplankton['CYC_ugDW/m3']
cyclopoid_babies = lake_erie_zooplankton['CYCIM_ugDW/m3']
daphnia = lake_erie_zooplankton['DAP_ugDW/m3']
non_daphnia = lake_erie_zooplankton['NDAP_ugDW/m3']
pred_cladocerans = lake_erie_zooplankton['PRED_ugDW/m3']
```

```python
fig, axs = plt.subplots(7,1,figsize=(14,36))
fig.tight_layout(pad=3.0)

sns.boxplot(x = year, y = calanoid_adults,hue = month, ax = axs[0])
axs[0].set_ylabel('μg/m$^{3}$')
axs[0].set_title('Calanoid Copepod Adults')

sns.boxplot(x = year, y = calanoid_babies,hue = month, ax = axs[1])
axs[1].set_ylabel('μg/m$^{3}$')
axs[1].set_title('Calanoid Copepod Copepodites')

sns.boxplot(x = year, y = cyclopoid_adults,hue = month, ax = axs[2])
axs[2].set_ylabel('μg/m$^{3}$')
axs[2].set_title('Cyclopoid Copepod Adults')

sns.boxplot(x = year, y = cyclopoid_babies,hue = month, ax = axs[3])
axs[3].set_ylabel('μg/m$^{3}$')
axs[3].set_title('Cyclopoid Copepod Copepodites')

sns.boxplot(x = year, y = daphnia,hue = month, ax = axs[4])
axs[4].set_ylabel('μg/m$^{3}$')
axs[4].set_title('Daphnia')

sns.boxplot(x = year, y = non_daphnia,hue = month, ax = axs[5])
axs[5].set_ylabel('μg/m$^{3}$')
axs[5].set_title('Non-Daphnia Herbivorous Cladocerans')

sns.boxplot(x = year, y = pred_cladocerans,hue = month, ax = axs[6])
axs[6].set_ylabel('μg/m$^{3}$')
axs[6].set_title('Predatory Cladocerans')
```

```python
plt.plot(lake_erie_zooplankton_month,zooplankton_med, color = 'pink', linewidth = 3)
plt.plot(lake_erie_zooplankton_month,zooplankton_25, linestyle = 'dashed', color = 'pink' )
plt.plot(lake_erie_zooplankton_month,zooplankton_75, linestyle = 'dashed', color = 'pink' )

plt.fill_between(lake_erie_zooplankton_month,zooplankton_25,zooplankton_75,color='pink', alpha=0.2)

plt.xlabel('Month')
plt.ylabel('Zooplankton Biomass (μg/m$^{3}$)')

plt.legend(['Zooplankton median'], loc = 'best')
plt.show()
```
