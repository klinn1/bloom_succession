```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import lognorm
from scipy import stats 
```

```python
mu = 3.0 
fig, ax = plt.subplots(1, 1)
x = np.linspace(lognorm.ppf(0, mu), 3)
#print(x)
#print(lognorm.pdf(x, mu))
print(lognorm.ppf(0.5,))
#ax.plot(x, lognorm.pdf(x, mu),'r-', lw=5, alpha=0.6, label='lognorm pdf')
plt.plot(x)
```

```python
plt.hist(lognorm.pdf(x, mu), normed=True)

# find minimum and maximum of xticks, so we know
# where we should compute theoretical distribution
xt = plt.xticks()[0]  
xmin, xmax = min(xt), max(xt)  
lnspc = np.linspace(xmin, xmax, len(x))
x = np.linspace(lognorm.ppf(0, mu), 3)

# normal distribution 
m, s = stats.norm.fit(lognorm.pdf(x, mu)) # mean and standard deviation  
pdf_g = stats.norm.pdf(lnspc, m, s) #theoretical values in interval  
plt.plot(lnspc, pdf_g, label="Norm") 
```

```python
psi = 1.2
phi = mu**psi

fig, ax = plt.subplots(1, 1)
x = np.linspace(lognorm.ppf(0, phi), 0.2)
print(x)
print(lognorm.pdf(x, phi))
ax.plot(x, lognorm.pdf(x, phi),'r-', lw=5, alpha=0.6, label='lognorm pdf')
```

```python
plt.hist(lognorm.pdf(x, phi), normed=True)

# find minimum and maximum of xticks, so we know
# where we should compute theoretical distribution
xt = plt.xticks()[0]  
xmin, xmax = min(xt), max(xt)  
lnspc = np.linspace(xmin, xmax, len(x))
x = np.linspace(lognorm.ppf(0, phi), 0.2)

# normal distribution 
m, s = stats.norm.fit(lognorm.pdf(x, phi)) # mean and standard deviation  
pdf_g = stats.norm.pdf(lnspc, m, s) # theoretical values interval  
plt.plot(lnspc, pdf_g, label="Norm")
```
